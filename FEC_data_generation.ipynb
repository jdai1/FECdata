## Committee ID to Candidate Name to Committee Name

```python
import pandas as pd
import os
```

```python
# candidate-committee linkage files
ccl_path = '/Users/Julian/Documents/CompSci/Numedi-Internship/FECdata/datasets/ccl'
for root, dirs, files in os.walk(ccl_path, topdown=True):
    files.sort()
    ccl_folder = files
ccl_folder.remove('.DS_Store')

print(ccl_folder)
```

```python
# candidate master files
cn_path = '/Users/Julian/Documents/CompSci/Numedi-Internship/FECdata/datasets/candidate_master'
for root, dirs, files in os.walk(cn_path):
    files.sort()
    cn_folder = files
cn_folder.remove('.DS_Store')
print(cn_folder)
```

```python
# committee master files
cm_path = '/Users/Julian/Documents/CompSci/Numedi-Internship/FECdata/datasets/committee_master'
for root, dirs, files in os.walk(cm_path, topdown=True):
    files.sort()
    cm_folder = files
print(cm_folder)
```

```python
# for CMTE_IDs without CAND_NAMEs, the CMTE_NAME is provided instead
for ccl_file, cn_file, cm_file in zip(ccl_folder, cn_folder, cm_folder):
    df_cn = pd.read_csv('/Users/Julian/Documents/CompSci/Numedi-Internship/FECdata/datasets/candidate_master/' + cn_file, delimiter='|', header=None, encoding='ISO-8859-1', low_memory=False)
    df_cn = df_cn[[0, 1]]
    df_cn.columns = ['CAND_ID', 'CAND_NAME']
    df_ccl = pd.read_csv('/Users/Julian/Documents/CompSci/Numedi-Internship/FECdata/datasets/ccl/' + ccl_file, delimiter='|', header=None, encoding='ISO-8859-1', low_memory=False)
    df_ccl = df_ccl[[3, 0]]
    df_ccl.columns = ['CMTE_ID', 'CAND_ID']
    print(df_ccl)
    df_cm = pd.read_csv('/Users/Julian/Documents/CompSci/Numedi-Internship/FECdata/datasets/committee_master/' + cm_file, delimiter='|', encoding='ISO-8859-1', header=None, low_memory=False)
    df_cm = df_cm[[0, 1]]
    df_cm.columns = ['CMTE_ID', 'CMTE_NAME']
    cand_name_list = []
    for i in range(0, df_ccl.shape[0]):
        row = df_ccl.iloc[i]
        cand_id = row['CAND_ID']
        cmte_row = df_cm[df_cm['CMTE_ID'] == row['CMTE_ID']]
        cand_row = df_cn[df_cn['CAND_ID'] == cand_id]['CAND_NAME']
        if cand_row.shape[0] > 0:
            cand_name = cand_row.iloc[0]
        else:
            cmte_name = cmte_row['CMTE_NAME']
            print(cmte_name)
            cand_name = cmte_name
        cand_name_list.append(cand_name)
    print(len(cand_name_list))
    dfp = pd.DataFrame()
    dfp['CMTE_ID'] = df_ccl['CMTE_ID']
    dfp['CAND_NAME'] = cand_name_list
    cfile = open('/Users/Julian/Documents/CompSci/Numedi-Internship/FECdata/datasets/ccl-cn/ccl-cn' + ccl_file[-6:], 'w+')
    
    cfile.write(dfp.to_csv(sep='|', index=False))
```
